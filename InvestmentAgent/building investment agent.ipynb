```python
import yfinance as yf
import pandas as pd
```

```python
lt = yf.Ticker("LT.NS")        # NSE symbol must end with .NS
data = lt.history(period="ytd") 

lt_data = pd.DataFrame(data)
print(lt_data.shape)
lt_data.tail()
```

```python
airtel = yf.Ticker("BHARTIARTL.NS")
data = airtel.history(period="ytd")

airtel_data = pd.DataFrame(data)
print(airtel_data.shape)
airtel_data.tail()
```

```python
nifty = yf.Ticker("^NSEI")
print(nifty.fast_info)
data = nifty.history(period="ytd")

nifty_data = pd.DataFrame(data)

print(nifty_data.shape)
nifty_data.tail()
```
